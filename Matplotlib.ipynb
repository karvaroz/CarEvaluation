# Importar librerias

```python
import matplotlib.pyplot as plt
```

## Line Plot

```python
# país 1
x = [2016, 2017, 2018, 2019, 2020, 2021]
y = [45, 46, 47, 48, 50, 51]
#país 2
# país 1
x2 = [2016, 2017, 2018, 2019, 2020, 2021]
y2 = [40, 41, 42, 43, 44, 45]

plt.figure(figsize=(10,6))
plt.plot(x, y, marker="o", linestyle="--", color="b", label="Colombia")
plt.plot(x2, y2, marker="d", linestyle="-", color="g", label="Argentina")
plt.xlabel("Años")
plt.ylabel("Población en Millones")
plt.title("Población (M) vs Años")
plt.legend(loc="best")
# plt.savefig("lineplot.png")
plt.yticks([45,48,51])
plt.show()
```

## Subplots

```python
fig, ax = plt.subplots(1,2, sharey = True)

ax[0].plot(x, y, color="b", label="Colombia")
ax[0].legend()
ax[1].plot(x2, y2, color="g", label="Argentina")
ax[1].legend()
plt.show()
```

## Bar Plots

```python
x = ["Argentina", "Colombia", "Perú"]
y = [40, 50, 33]

plt.bar(x, y)
plt.xlabel("Países")
plt.ylabel("Población")
plt.title("Gráfico de Barra")
plt.show()
```

## Pie Chart

```python
plt.pie(y, labels = x)
plt.show()
```

## Histogram

```python
edades = [15, 16, 17, 20, 21, 22, 23, 24, 25, 26, 30, 31, 32, 35]
bins = [15, 20, 25, 30, 35]

plt.hist(edades, bins, edgecolor="white")
plt.show()
```

## Boxplots

```python
edades = [15, 16, 17, 20, 21, 22, 23, 24, 25, 26, 30, 31, 32, 48]

plt.boxplot(edades)
plt.show()
```

## Scatterplot

```python
a = [1, 2, 3, 4, 5, 4, 3, 2, 2, 4, 5, 6, 7]
b = [7, 2, 3, 5, 5, 7, 3, 2, 1, 4, 6, 3, 2]

plt.scatter(a, b)
plt.show()
```
